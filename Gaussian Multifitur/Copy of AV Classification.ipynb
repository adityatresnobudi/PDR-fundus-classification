```python
# NYOBA PCA-SVM UBAH NILAI GAMMA + C
X, y = df2.iloc[:, 0:15].values, df2.iloc[:, 105].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0, shuffle=True)
```

```python
pca = PCA()
X_train_pca = pca.fit_transform(X_train)
pca.explained_variance_ratio_
```

```python
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
#from sklearn.linear_model import LogisticRegression
#model = LogisticRegression()

#model = KNeighborsClassifier(n_neighbors=10, p=2, metric='minkowski')
#model = GaussianNB()
model = SVC(kernel='rbf', gamma = 1/10, C=1, random_state=0)
model = model.fit(X_train_pca, y_train)

print('Training accuracy:', model.score(X_train_pca, y_train))
print('Test accuracy:', model.score(X_test_pca, y_test))
```
